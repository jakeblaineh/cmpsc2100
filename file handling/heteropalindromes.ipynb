```python
def heteropalindromes(words_dictionary):
    with open(words_dictionary, 'r') as words:
        words_set = set()
        for word in words:
            words_set.add(word.strip())
    
    heteropalindromes = set()
    for word in words_set:
        if word[::-1] in words_set:
            if (word[::1], word) not in heteropalindromes and (word, word[::-1]) not in heteropalindromes:
                heteropalindromes.add((word, word[::-1]))
        
    return heteropalindromes
```
